# Creating probablilty model

```python
inputs = [col for col in list(header.columns) if header.loc[1,col]=='N']
```

```python
inputs
```

```python
len(inputs)
```

```python
x = np.array(df[inputs])
y = np.array(df['lift'])
```

```python
states = Input(shape=(26,))
l1 = Dense(units = 13, activation='linear')(states)
l2 = Dense(units = 6,activation='linear')(l1)
l3 = Dense(units = 3,activation='linear')(l2)
q_value = Dense(units = 1,activation='sigmoid')(l3)
ann_model = Model(inputs=states,outputs=q_value)
ann_model.compile(loss='binary_crossentropy',optimizer=Adam(lr=.000001))
```

```python
ann_model.fit(x,y,epochs=10000,batch_size=100)
```

```python
ann_model.save('../Models/prob_model.h5')
```

# Static evaluation on dev and val

```python
prob_model = load_model('../Models/prob_model.h5')
pos_return_model = load_model('../Models/pos_return_model.h5')
neg_return_model = load_model('../Models/neg_return_model.h5')
```

```python
prob_inputs = [col for col in list(header.columns) if header.loc[1,col]=='N']
pos_return_inputs = [col for col in list(header.columns) if header.loc[2,col]=='N']
neg_return_inputs = [col for col in list(header.columns) if header.loc[3,col]=='N']
```

```python
df['expected_probability'] = prob_model.predict(df[prob_inputs])
```

```python
df['expected_pos_return'] = pos_return_model.predict(df[pos_return_inputs])
```

```python
df['expected_negative_return'] = neg_return_model.predict(df[neg_return_inputs])
```

```python
df['expected_pos_return'] = df['expected_pos_return'].apply(lambda x: np.exp(x)-1 if x >0 else 0)
df['expected_negative_return'] = df['expected_negative_return'].apply(lambda x : np.exp(x)-1 if x <0 else 0)
```

```python
df['expected_return'] = df['expected_probability']*df['expected_pos_return']+(1-df['expected_probability'])*df['expected_negative_return']
```

```python
df['expected_return'] = df['expected_return'].apply(lambda x: np.log(1+x))
```

```python
df['expected_lift'] = df['expected_return'].apply(lambda x: 1 if x > 0 else 0)
```

```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
import pandas as pd
import numpy as np
import pickle
```

```python
stats = sqlContext.sql("""select symbol,substr(cast(date as string),1,7) as month,
count(*) as trades,
sum(lift) as actual_profits,
sum(expected_lift) as predicted_profits,
sum(case when lift = 0 and expected_lift = 0 then 1 else 0 end) as loss_correct_predicted,
sum(case when lift =1 and expected_lift = 1 then 1 else 0 end) as profit_correct_predicted
from data
group by 1,2
""").toPandas()
```
